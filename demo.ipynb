```python
import os
import sys
import json

# Ensure the local modules can be imported
module_path = os.path.abspath(os.path.join('.'))
if module_path not in sys.path:
    sys.path.append(module_path)

from data_fetcher import fetch_polymarket_event
from llm_brain import analyze_market_event_multi_agent
```

## Phase 1: Data Ingestion

```python
# Example parameters (Feel free to change these)
POLYMARKET_SLUG = "will-crude-oil-cl-hit-by-end-of-march"

print(f"Fetching baseline event data for {POLYMARKET_SLUG}...")
event_info = await fetch_polymarket_event(POLYMARKET_SLUG)

print("\n--- Event Data (Phase 1 Output) ---")
print(json.dumps(event_info, indent=2))
```

## Phase 2: LLM Synthesis Layer

```python
from dotenv import load_dotenv
load_dotenv() # Load variables from .env file

if not os.environ.get("GEMINI_API_KEY"):
    print("\nWARNING: GEMINI_API_KEY is not set.")
    print("Please create a file named `.env` in this directory containing: GEMINI_API_KEY=your-actual-key")

print("Activating Hedge Fund Multi-Agent Debate...\n")
allocation_plan = await analyze_market_event_multi_agent(POLYMARKET_SLUG)

print("\n--- Multi-Agent Allocation Plan (Phase 2 Output) ---")
print(json.dumps(allocation_plan, indent=2))
```

## Phase 3: Trade Execution Layer
